```python
import os
import pandas as pd
import matplotlib. pyplot as plt
```

## Merge all references into single DataFrame

```python
df = {
    'irrelevant': pd.read_csv(os.path.join('data', 'references', 'review_176995_irrelevant_csv_20220512005557.csv')),
    'excluded': pd.read_csv(os.path.join('data', 'references', 'review_176995_excluded_csv_20220512005107.csv')),
    'included': pd.read_csv(os.path.join('data', 'references', 'review_176995_included_csv_20220512024607.csv'))
}
```

```python
for decision in df.keys():
    df[decision]['decision'] = decision
```

```python
df['all'] = pd.concat(df.values()).reset_index(drop=True)
```

Replace NaN title with empty string to allow easy searching:

```python
df['all'].loc[df['all'].Title.isnull(), 'Title'] = ''
```
